```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt("eigenvectors.dat")
```

```python
r,c = np.shape(data)
```

```python
i = 1
while(i<c):
    plt.plot(data[:,0],np.abs(data[:,i]))
    #plt.ylim([-0.1,0.6])
    #plt.show()
    i = i + 1
plt.xlabel("No. of training data points",fontsize=15)
plt.ylabel("$|$HLDA Coefficient$|$",fontsize=15)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.title("CLN025 in implicit solvent",fontsize=15)
plt.savefig('Chignolin_implicit.jpg',bbox_inches="tight")
```
