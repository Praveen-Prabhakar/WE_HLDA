```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
np.set_printoptions(threshold=np.inf)
import math
import westpa.cli.tools.w_ipa as w_ipa
from westpa.analysis import BasicMDTrajectory
from scipy import stats
w = w_ipa.WIPI()

# At startup, it will load or run the analysis schemes specified in the configuration file (typically west.cfg)
w.main()
w.interface = 'matplotlib'
```

```python
total_iteration =  300

dt = 0.2 #frequency at which pcoord data is saved

sink = 0.031 #target state

time_unit = 'ps'

#-----------------------------------------------------------------------------#
#--NO NEED TO MODIFY ANYTHING BELOW UNLESS YOU WANT TO CHANGE THE ALGORITHM---#
#-----------------------------------------------------------------------------#
it = np.zeros(total_iteration) #array to store flux data
time = np.zeros(total_iteration) #array to store time data

force_eval = 0.0

for i in range(total_iteration):
    #find out which iteration you are in
    w.iteration = i+1
    
    #get an array of all the segments you have in this iteration
    l = w.current.pcoord
    
    #get the weight of all the segments in this iteration
    wts = w.current.weights
    
    #size of each segment
    tau = len(l[0])
    
    #calculate total computational time spent (size * number of segment * time unit)
    force_eval += (tau-1)*len(l)*dt
    print('total simulation time = ', force_eval, time_unit)
    
    #run loop to see which trajectories crossed the sink
    for j in range(len(l)):
        if l[j][0] < sink and l[j][tau-1] >= sink:
            print(l[j][0])
            print(l[j][tau-1])
            #add its weight to the flux array
            it[i] += wts[j]
            
     
    
    try: 
        time[i] = (i/sum(it))*(tau-1)*dt #compute first passage time as time/flux
    except: 
        time[i] = 0.0 #take care of iterations when no segment reached the sink
```

```python
plt.plot(1/time,'-r')
#plt.ylim(0,1200)
plt.xlabel('Iterations',fontsize=14)
plt.ylabel('Flux per unit time (ps$^{-1}$)',fontsize=14)
plt.title('ALD in vacuum',fontsize=14)
#plt.savefig('Flux_Ala_vacc.jpg',bbox_inches="tight",dpi = 300)
```

```python
#Calculation of simulation time
from westpa.analysis import Run
run = Run.open("west.h5")
num_iter = run.num_iterations
resampling_time = 2 #in ps
st = np.zeros((num_iter,1))
i = 1
while(i<=num_iter):
    iter_current = run.iteration(i)
    num_walkers = iter_current.num_walkers
    st[i-1] = num_walkers*resampling_time
    i = i + 1
ST = sum(st)/1000
print('total simulation time = ',ST,"ns")
run.close()
run.h5file
```
